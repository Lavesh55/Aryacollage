# Q.19

```python
import matplotlib.pyplot as plt
```

```python
department_avg_salary = df.groupby('Department')['Salary'].mean()
plt.figure(figsize=(8, 6))
department_avg_salary.plot(kind='bar', color='skyblue')
plt.title('Department-wise Average Salary')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df2=pd.read_csv("Copy of Screenshot 2025-03-28 072452(1).csv")
df2
```

# Q.21

```python
department_count = df2['Department\nHR'].value_counts()
plt.figure(figsize=(8, 6))
department_count.plot(kind='bar', color='skyblue')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Q.22

```python
high_salary_employees = df2[df2['Salary\n50000'] > 60000]
print("Employees with salary greater than 60,000:")
print(high_salary_employees)
```

# Q 23

```python
average_salary_by_department = df2.groupby('Department\nHR')['Salary\n50000'].mean()
plt.figure(figsize=(8, 6))
average_salary_by_department.plot(kind='line', marker='o', color='green')
plt.title('Salary Trend by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
